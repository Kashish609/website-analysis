### what patterns or trends can you observe in website sessions and users, over time.

```python
plt.figure(figsize = (10,5))
sns.set(style = "whitegrid")
data.groupby('datehour')[['users', 'sessions']].sum().plot(ax=plt.gca())
plt.title("sessins ans users over time")
plt.xlabel("datehour")
plt.ylabel("Count")
```

### which marketting channel brought the highest number of users to the website, and how can we use this insights to improve traffic from other sources?

```python
result = data.groupby('channel group')['users'].sum().reset_index()
result
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="channel group", y="users", estimator=np.sum, palette="viridis") #onlt for bar plot
plt.title("Total Users by Channel")
plt.xlabel("Channel Group")
plt.ylabel("Total Users")
plt.xticks(rotation="vertical")
plt.show()
```

### which channel has the highest average engagement time, what does that tells us about user behaviour and content effectiveness.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="channel group", y="Average engagement time per session", estimator=np.mean, palette="magma") #onlt for bar plot
plt.title("highest average engagement")
plt.xlabel("Channel Group")
plt.ylabel("Average engagement time per session")
plt.xticks(rotation="vertical")
plt.show()
```

### how does engagement rate vary across different traffic channels?

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x = "channel group", y = "Engagement rate", palette="coolwarm")
plt.title("engagement rate distribution")
plt.xlabel("channel")
plt.ylabel("rate")
plt.xticks(rotation="vertical")
```

### which channels are driving more engaged sessions comparing the non engaged ones, and what can improve engagement in underperforming channels.

```python
# Step 1: Create a new column for non-engaged sessions
data['non_engaged_sessions'] = data['sessions'] - data['engaged sessions']

# Step 2: Group by channel group and sum the sessions
combined_data = data.groupby('channel group')[['engaged sessions', 'non_engaged_sessions']].sum()

# Step 3: Plot it
combined_data.plot(kind='bar', figsize=(8,5), color=['#66b3ff', '#ff9999'])
plt.title('Engaged vs Non-Engaged Sessions by Channel')
plt.xlabel('Channel Group')
plt.ylabel('Number of Sessions')
plt.legend(title='Session Type')
plt.tight_layout()
plt.show()
```

### traffic by hour and channels 

```python
data['Hour'] = data['datehour'].dt.hour
# Group and prepare data for the heatmap
heatmap_data = data.groupby(['Hour', 'channel group'])['sessions'].sum().unstack().fillna(0)

#  Create a cleaner, prettier heatmap
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5, annot=True, fmt='.0f')
plt.title("ðŸ’¡ Traffic by Hour and Channel", fontsize=14)
plt.xlabel("Channel Group", fontsize=12)
plt.ylabel("Hour of Day", fontsize=12)
plt.tight_layout()
plt.show()
```
